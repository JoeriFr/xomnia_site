```python
CONVERT_DICT = {
    "datetime": int,
    "lat": float,
    "lon": float,
    "spd_over_grnd": float,
    "true_course": float,
    "mag_variation": float,
}

df = pd.read_csv("data/raw_messages.csv")
df["raw_message"] = df.raw_message.str.replace(r"[^0-9a-zA-Z.,]", "", regex=True)
df["raw_message"] = df["raw_message"].str.split(",")
x = pd.DataFrame(
    df["raw_message"].to_list(),
    columns=[
        "status",
        "lat",
        "lat_dir",
        "lon",
        "lon_dir",
        "spd_over_grnd",
        "true_course",
        "datestamp",
        "mag_variation",
        "mag_var_dir",
    ],
)
result = pd.concat([df, x], axis=1, join="inner").drop("raw_message", axis=1)
result = result.astype(CONVERT_DICT)
result["datetime_iso"] = pd.to_datetime(result["datetime"], unit="s")

result.to_pickle("data/clean_messages.pkl")
result.to_csv("data/clean_messages.csv")
# print(result.infer_objects().dtypes)
print(result.dtypes)
print(result)
# print(result["status"])
# exit()
# df[["status","lat","lat_dir","lon","lon_dir","spd_over_grnd","true_course","datestamp","mag_variation","mag_var_dir"]] = pd.DataFrame(df["raw_message"].to_list(), index= df.index)
# print(df2)
```
